```python
from AAPI_code.Collage_generator import collage_generator
from AAPI_code.coco_converter import coco_converter
from AAPI_code.Collage_generator.utils import *
```

# collage_generator

```python
col_gen = collage_generator(label_list= [], 
                            canvas_size=(2000,2000),
                            example_image = "/content/AAPI_code/data/vignettes/background/11000_16000_0_background.png",
                            gaussian_noise_constant = 5,
                            cluster_size = (1000,1000),
                            patience = 25
                            )
col_gen.import_images_from_directory("/content/AAPI_code/data/vignettes")
```

# Generation Example

```python
collage,mask, dictionary = col_gen.generate(item_num= 10,
                                            ratio_dict={"cluster":0.1, "artery": 0.5, 'arteriole': 0.4},
                                            background_color = False,
                                            format = 'COCO')
visualize_result(collage,mask, dictionary)# visualize can take new version mask as well
```
